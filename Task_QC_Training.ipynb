````python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
````

# **Create a Quantum Circuit with a Hadamard Gate**

- Write a Qiskit program to create a single-qubit circuit.
- Apply a Hadamard gate to the qubit.
- Measure the qubit and visualize the result using a histogram.

````python
circuit = QuantumCircuit(1, 1)
circuit.h(0)
circuit.measure(0, 0)

result = AerSimulator().run(circuit).result()
statistics = result.get_counts()

display(plot_histogram(statistics))
````

# **Implement a Quantum Coin Flip**
- Simulate a fair coin flip using a quantum circuit.
- Print the probability of getting Heads or Tails.

````python
circuit = QuantumCircuit(1, 1)
circuit.h(0)
circuit.measure(0, 0)

result = AerSimulator().run(circuit).result()
statistics = result.get_counts()

total = sum(statistics.values())

print(f"Chances of getting Heads: {(statistics["0"]/total) * 100:.2f} %\n"
      f"Chances of getting Tails: {(statistics["1"]/total) * 100:.2f} %")

display(plot_histogram(statistics))
````

# **Implement a Bell State (Entanglement)**
- Create a two-qubit circuit.
- Apply a Hadamard gate to the first qubit and a CNOT gate between the two qubits.
- Measure both qubits and verify if they are always correlated.

````python
circuit = QuantumCircuit(2, 2)

circuit.h(0)
circuit.cx(0, 1)

circuit.measure(0, 0)
circuit.measure(1, 1)

result = AerSimulator().run(circuit).result()
statistics = result.get_counts()

display(plot_histogram(statistics))
````

### **How does superposition affect the information processing capabilities of a quantum computer?**

The superposition state allows QCs to process data in a simultaneous way, accounting multiple options parallelly and executing an algorithm on all options in a single step. This gives QC comparatively much faster computing time than CC, especially for problems involving huge data or complex computations.

### **How do quantum gates manipulate interference to solve problems faster than classical computers?**

Quantum computers manipulate interference by controlling the phase of quantum states. Using such gates as 'T' and 'S' gates to modify the phase, which in turn affects the interference of the quantum states.

This phase manipulation leads to Destructive Interference(DI) or Constructive Interference(CI). CI occurs when the amplitues of the states align, maximizing the amplitude. On the other hand, DI occurs when the amplitues cancel each other out, reducing the amplitude.

Algorithms like 'Grovers Algorithm' for instance uses this concept to get the correct answer by applying phase shifts. This over multiple iterations increases the amplitude of the correct answer(Constructive Interference) while the incorrect ones' amplitudes decreases (Destructive Intereference).
